```python
def biseccion(f,a,b,tol):
    
    if f(a)*f(b)>0:
        return 'Error f(a) y f(b), deben tener signos contrarios'
    
    p=(a+b)/2
    
    n=0
    
    while abs(f(p))>tol:
        
        if f(a)*f(p)<0:
            b=p
        if f(p)*f(b)<0:
            a=p
            
        p=(a+b)/2
        n+=1
        
        #print('n=',n,'   f(p):',f(p))
        
    print('Raiz encontrada en', n,'iteraciones')
        
    return p
```

```python
def sen(grados):
    return math.sin(math.radians(grados))
def cos(grados):
    return math.cos(math.radians(grados))
def tan(grados):
    return math.tan(math.radians(grados))


def julian_day(fecha):
    J2000 = 2451545
    day_r = pd.to_datetime('2000-01-01 12:00')
    dt = (fecha - day_r).days
    return (J2000 + dt)
```

\begin{equation}
\underbrace{
\left(
\begin{array}{ccc}
cos(\epsilon) & -sen(\epsilon) & 0 \\
sen(\epsilon) & cos(\epsilon) & 0 \\
0 & 0 & \frac{1}{cos(\lambda)cos(\beta)}
\end{array}
\right)
}_{A}
\cdot
\underbrace{
\left(
\begin{array}{c}
sen(\delta) \\
\alpha_{n} \\
\alpha_{d} \\
\end{array}
\right)
}_{x}
=
\underbrace{
\left(
\begin{array}{c}
sen(\beta) \\
cos(\beta)sen(\lambda) \\
1 \\
\end{array}
\right)
}_{b}
\end{equation}

$$\alpha = tan^{-1}\left( \frac{\alpha_{n}}{\alpha_{d}}\right)$$

```python
def row_lamb(A,i,l):
    B=np.copy(A)
    B=B.astype('float64')
    
    B[i,:] = B[i,:]*l
    return B

def row_comb(A,i,j,l):
    B=np.copy(A)
    B=B.astype('float64')
    
    B[i,:] += row_lamb(A,j,l)[j,:]
    return B

def row_swap(A,i,j):
    B=np.copy(A)
    B=B.astype('float64')
    
    B[[i,j]]=B[[j,i]]
    
    return B

def Gauss(A):
    B=np.copy(A)
    #B=B.astype('float64')
    
    m=len(B[:,0]) #número de filas
    n=len(B[0,:]) #número de columnas
    
    #para matrices no cuadradas
    k=min([m,n]) 
    
    for j in range(k): #Corremos sobre las columnas
        
        for i in range(j,m): #Corremos sobre las filas
            if B[i,j]!=0: #Encontrar el pivote
                break
        
        B=row_swap(B,i,j)
        
        if B[j,j]==0: #Si no hay pivote pasar a la siguiente columna
            continue
            
        for i in range(j+1,m): #corremos sobre el resto de filas
            B=row_comb(B,i,j,-B[i,j]/B[j,j])
            
    
    if B[j,j]!=0:
        B=row_lamb(B,j,1/B[j,j])
        
    return B


```

```python
def Gauss_Solve(A,b):
    
    m=len(A[:,0]) #filas
    n=len(A[0,:]) #columnas
    
    G=Gauss(np.c_[A,b])
    
    Ap=G[:,0:n]
    bp=G[:,-1]
    
    x=np.zeros(m)
    
    for i in range(n-1,0-1,-1): #Desde la última hasta la primera
        if Ap[i,i]==0:
            print('¡Alerta! No hay pivote en la fila %s x[%s] será tomado como un parámetro' %(i,i))
            x[i]=1. #Puede ser de su elección
            continue
        #j=i+1:n   
        x[i]=(bp[i]-sum(Ap[i,i+1:n]*x[i+1:n]))/Ap[i,i]
        
    return x

```
